```python
# Script Automatizado Nexus Éter para Geração e Sincronização de Legendas (SRT)
# Requisitos: Google Colab com GPU habilitada

# Instalação das Dependências
!pip install git+https://github.com/openai/whisper.git
!pip install torch torchvision torchaudio
!pip install ffsubsync
!sudo apt update && sudo apt install ffmpeg

print('Instalação de dependências concluída.')

# Upload do Vídeo
from google.colab import files
print('Faça o upload do vídeo:')
uploaded = files.upload()

video_file = list(uploaded.keys())[0]
print(f'Vídeo recebido: {video_file}')

# Extração do Áudio com FFmpeg
audio_file = video_file.replace('.mp4', '.wav')
!ffmpeg -i "$video_file" -q:a 0 -map a "$audio_file"
print(f'Áudio extraído: {audio_file}')

# Transcrição com Whisper
print('Iniciando a transcrição...')
!whisper "$audio_file" --model large --language English --task transcribe --output_format srt --device cuda
print('Transcrição concluída.')

# Ajuste de Sincronização com ffsubsync
print('Sincronizando a legenda...')
synced_srt = audio_file.replace('.wav', '_synced.srt')
!ffsubsync "$video_file" -i "$audio_file.srt" -o "$synced_srt"
print(f'Legenda sincronizada: {synced_srt}')

# Download da Legenda Sincronizada
print('Pronto para download:')
files.download(synced_srt)
```
